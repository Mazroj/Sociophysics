```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_global_beta_vs._mu_p30.csv')
df1 = df.to_numpy().tolist()
df1 = np.array(df1)
```

```python
df1
```

```python
plt.figure(figsize =(12,9))
plt.imshow(df1,origin = 'lower' )
plt.colorbar()
plt.xlabel(r"$\mu$/p")
plt.ylabel(r"$\beta$")
plt.title(r"Intensity $\beta$ vs. Tolerance $\mu$ - Global Network")
plt.show()
#plt.savefig("global2_beta_vs_mu_n100_p10.png")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_squared_beta_vs._mu.csv')
df1 = df.to_numpy().tolist()
df1 = np.array(df1)/100
```

```python
df1
```

```python
plt.figure(figsize =(12,9))
plt.imshow(df1,origin = 'lower' )
plt.colorbar()
plt.xlabel(r"$\mu$/p")
plt.ylabel(r"$\beta$")
plt.title(r"Intensity $\beta$ vs. Tolerance $\mu$ - Square Network")
#plt.show()
plt.savefig("squared_beta_vs_mu.png")
```

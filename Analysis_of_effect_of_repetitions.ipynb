# Effect of repetition

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import discrete_opinion as do
```

## Analyzing data

```python
def L2(lista):
    l2 = np.sqrt((np.sum((lista)**2)/(len(lista)**2)))
    return l2
```

## Test Run

```python
import time
ob1 = do.Global(size = 100,options = 20)
time_list = []
for t in range(0,20):
    a = time.time()
    rep = []
    for r in range(0,50):
        lisq = ob1.simul_global_evol(t,0.4,10**5,18)
        rep.append(lisq)
    lis = np.mean(rep, axis = 0)
    b = time.time()
    print("Time: ", b-a)
    time_list.append(b-a)
```

```python
plt.plot(np.array(time_list)/50,'.-')
plt.xlabel(r"$\mu$")
plt.ylabel("t")
plt.title("Time vs. Tolerance")
plt.show()
```

## Repetitions

```python
import time
ob1 = do.Global(size = 100,options = 20)
smax_per_tolerance = np.empty(20)
sprop_per_tolerance = np.empty(20)
smax_desv = np.empty(20)
for t in range(0,20):
    smax_list = np.empty(50)
    sprop_list = np.empty(50)
    for r in range(0,50):
        lisq = ob1.simul_global_evol(t,0.4,10**5,18)
        smax, sprop = ob1.most_frequent_value(lisq,18)
        smax_list[r] = smax
        sprop_list[r] = sprop
    smax_per_tolerance[t] = np.mean(smax_list)
    sprop_per_tolerance[t] = np.mean(sprop_list)
    smax_desv[t] = np.std(smax_list)
```

```python
plt.plot(smax_per_tolerance-sprop_per_tolerance, ".-")
plt.errorbar(np.arange(0,20,1),(smax_per_tolerance-sprop_per_tolerance),smax_desv)
```

### 50 repetitions

```python
databrut = []
for i in range(50):
    df = pd.read_csv("analysis_of_effect_repetitions_data/data_global_intensity_1_{:2d}".format(i))
    df1 = df.to_numpy().tolist()
    databrut.append(df1)

datamean = []
desv = []

for j in range(100):
    men = np.mean(np.transpose(databrut)[0][j])
    des = np.std(np.transpose(databrut)[0][j])
    datamean.append(men)
    desv.append(des)

```

```python
plt.plot(np.linspace(0,1,100),np.array(datamean)/1000,'.-')
```

```python
plt.plot(np.linspace(0,1,100),np.array(desv)/1000,'.-')
plt.xlabel(r'$\mu/p$')
plt.ylabel(r'$\sigma$')
plt.title('Standard Deviatiation of a serie of simulations n_1000_p_100_beta_0.1')
plt.show()
```

```python
L2dat = L2(np.array(datamean))
print("L2 error: ", L2dat)
```

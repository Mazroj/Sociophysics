```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
databrut = []
for i in range(50):
    df = pd.read_csv("data_global_intensity_1_{:2d}".format(i))
    df1 = df.to_numpy().tolist()
    databrut.append(df1)

datamean = []
desv = []
for j in range(100):
    men = np.mean(np.transpose(databrut)[0][j])
    des = np.std(np.transpose(databrut)[0][j])
    datamean.append(men)
    desv.append(des)
    
```

```python
np.transpose(databrut)[0][2]
```

```python
plt.plot(np.linspace(0,1,100),np.array(datamean)/1000,'.-')
```

```python
plt.plot(np.linspace(0,1,100),np.array(desv)/1000,'.-')
plt.xlabel(r'$\mu/p$')
plt.ylabel(r'$\sigma$')
plt.title('Standard Deviatiation of a serie of simulations n_1000_p_100_beta_0.1')
plt.show()
```
